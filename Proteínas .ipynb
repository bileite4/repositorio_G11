```python
from Bio.Blast import NCBIXML
from Bio.Blast import NCBIWWW
from Bio import SeqIO
```

```python
# Proteína INS
# Ler a sequencia de aminoácidos FASTA
seq_record = SeqIO.read(open("INS_proteina.txt"),format="fasta")

# Executar o BLAST usando a sequencia
result_handle = NCBIWWW.qblast("blastp", "swissprot", seq_record.format("fasta"))

# Gravar ficheiro
gravar_ficheiro=open("myblast.xml","w")
gravar_ficheiro.write(result_handle.read())
gravar_ficheiro.close()

# Analisar os resultados
abrir_ficheiro=open("myblast.xml")
blast_records = NCBIXML.parse(abrir_ficheiro)
for blast_record in blast_records:
    for alignment in blast_record.alignments:
        for hsp in alignment.hsps:
            if hsp.expect < 1e-5: # threshold para e-value
                print('----------Alignment----------')
                print('Accession number:',alignment.accession)
                print('Sequence:',alignment.title)
                print("Length:",alignment.length)
                print("e_value:",hsp.expect)
```
