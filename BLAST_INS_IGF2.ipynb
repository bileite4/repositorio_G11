```python
from Bio.Blast import NCBIWWW
from Bio import SeqIO
from Bio.Blast import NCBIXML
```

```python
# GENE INS
# Ler a sequencia de arquivo FASTA
seq_record = SeqIO.read(open("INS.fasta"),format="fasta")

# Executar o BLAST usando a sequencia
result_handle = NCBIWWW.qblast("blastn", "nt", seq_record.format("fasta"))

# Analisar os resultados
sequences_found=[]
blast_records = NCBIXML.parse(result_handle)
for blast_record in blast_records:
    for alignment in blast_record.alignments:
        for hsp in alignment.hsps:
            if hsp.expect < 1e-50: # threshold para e-value
                # para evitar repeties
                if alignment.accession not in [sequence["accession"] for sequence in sequences_found]:
                    sequences_found.append({"accession": alignment.accession, "title": alignment.title, "length": alignment.length, "e-value": hsp.expect})
                    for sequence in sequences_found:
                        print('----------Alignment----------')
                        print('Accession number:', sequence["accession"])
                        print('Sequence:', sequence["title"])
                        print("Length:", sequence["length"])
                        print("e_value:", sequence["e-value"])
```
